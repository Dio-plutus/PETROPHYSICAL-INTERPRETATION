```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import lasio
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
import sklearn.model_selection as ms
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
def log_plot(logs):
    logs = logs.sort_values(by='DEPTH')
    top = logs.DEPTH.min()
    bot = 3508.858
    
    f, ax = plt.subplots(nrows=1, ncols=5, figsize=(12,8))
    ax[0].plot(logs.DT, logs.DEPTH, color='red')
    ax[1].plot(logs.GR, logs.DEPTH, color='black')
    ax[2].plot(logs.NPHI, logs.DEPTH, color='blue')
    ax[3].plot(logs.RHOB, logs.DEPTH, color='c')
    ax[4].plot(logs.RT, logs.DEPTH, color='m')
    
    for i in range(len(ax)):
        ax[i].set_ylim(top,bot)
        ax[i].invert_yaxis()
        ax[i].grid()
        
    ax[0].set_xlabel("DT (us/ft)")
    ax[0].set_xlim(logs.DT.min(),logs.DT.max())
    ax[0].set_ylabel("Depth(ft)")
    ax[1].set_xlabel("GR (API)")
    ax[1].set_xlim(logs.GR.min(), logs.GR.max())
    ax[2].set_xlabel("NPHI (v/v)")
    ax[2].set_xlim(logs.NPHI.min(),logs.NPHI.max())
    ax[3].set_xlabel("RHOB (g/cm3)")
    ax[3].set_xlim(logs.RHOB.min(),logs.RHOB.max())
    ax[4].set_xlabel("RT (ohm.m)")
    ax[4].set_xlim(logs.RT.min(),logs.RT.max())
    
    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]);
    ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[4].set_yticklabels([]) 
    
    f.suptitle('Well: 15-9-F-12', fontsize=14,y=0.94)
```

```python
log_plot(dt)
```

```python
def train_data(df, col, start, middle, end):
    
    df1 = df
    print(df1.head())
    test = df.loc[start:middle]
    train = df.loc[(middle+1): end]
    
    train = train.dropna()
    train = train.reset_index(drop=True)
    
    labels = train[col]
    
    train = train.drop(col, axis=1)
    test = test.drop(col, axis=1)
    
    features = np.array(train)
    test = np.array(test)
    
    ## Randomly sample cases to create independent training and test data
    np.random.seed(9988)
    indx = range(features.shape[0])
    indx = ms.train_test_split(indx, test_size = 0.4)
    x_train = features[indx[0],:]
    y_train = np.ravel(labels[indx[0]])
    x_test = features[indx[1],:]
    y_test = np.ravel(labels[indx[1]])
    
    ## Randomly sample cases to create independent validation and test data
    np.random.seed(9988)
    indx1 = range(x_test.shape[0])
    indx1 = ms.train_test_split(indx1, test_size = 0.5)
    x_test1 = x_test[indx1[0],:]
    y_test1 = np.ravel(y_test[indx1[0]])
    x_test2 = x_test[indx1[1],:]
    y_test2 = np.ravel(y_test[indx1[1]])
    
    xg = xgb.XGBRegressor(boosting_type='dart', learning_rate=0.01, n_estimators=4000, max_depth=15, base_score=0.7, random_state=42)
    eval_set = [(x_test1, y_test1)]
    xg.fit(x_train, y_train, eval_set=eval_set, early_stopping_rounds=200)
    
    lg = lgb.LGBMRegressor(boosting_type='gbdt', learning_rate=0.01, n_estimators=4000, max_depth=15)
    eval_set = [(x_test1, y_test1)]
    lg.fit(x_train, y_train, eval_set=eval_set, early_stopping_rounds=200)

    #lg = RandomForestRegressor(n_estimators=100, max_depth=15, min_samples_split=2, random_state=42)
    #lg.fit(x_train, y_train)
    
    xg_pred = xg.predict(x_test2)
    lg_pred = lg.predict(x_test2)
    
    sample_xg = xg_pred[:20]
    sample_lg = lg_pred[:20]
    
    rmse_xg = mean_squared_error(y_test2, xg_pred)
    rmse_lg = mean_squared_error(y_test2, lg_pred)
    
    print(f'The RMSE of the {col} using XGBOOST is {mean_squared_error(y_test2, xg_pred)} and The RMSE of the {col} using a LightGBM is {mean_squared_error(y_test2, lg_pred)}')
    print(f'Comparing the test results of the two models')
    print('-----------------------------------------------')
    print(f'Real {col}  ----------------- XGBoost ----------------- LightGBM')
    for i in range(20):
        print(f'{y_test2[i]} -------- {sample_xg[i]} -------- {sample_lg[i]}')
        print('---------------------------')
    
    if rmse_xg < rmse_lg:
        model = xg
    else:
        model = lg
        
    missing_pred = model.predict(test)
    
    df1[col].loc[start:middle] = missing_pred
    print(model.feature_importances_)
    
    return df1
```

```python
new_df = train_data(dt, 'NPHI', 0, 3137, 21017)
```
